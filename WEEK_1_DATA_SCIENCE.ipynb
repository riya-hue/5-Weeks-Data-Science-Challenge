ðŸ”¹ DAY 3-4: Pandas + Matplotlib + Seaborn

âœ… 4. Working with Pandas

```python
import pandas as pd

#create dataframe
data={
    'Name':['A','B','C'],#comma is important
    'Marks':[88,76,93]
}
df = pd.DataFrame(data)
print(df)

# Describe statistics
print(df.describe())
```

âœ… 5. Matplotlib & Seaborn Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line plot
plt.plot([1, 2, 3], [4, 5, 1])
plt.title("Line Plot")
plt.show()

# Seaborn bar plot
sns.barplot(x='Name', y='Marks', data=df)
plt.title("Marks of Students")
plt.show()
```

ðŸ”¹ DAY 5-6: Statistics & Probability

âœ… 6. Basic Statistics

```python
data = [10, 20, 30, 30, 40]

# Mean, Median, Mode
from statistics import mean, median, mode

print("Mean:", mean(data))
print("Median:", median(data))
print("Mode:", mode(data))
```

âœ… 7. Probability Distributions

```python
import scipy.stats as stats

# Normal Distribution: mean=0, std=1
x = np.linspace(-5, 5, 100)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("Normal Distribution")
plt.show()
```
